```python
## 下面是低维数据处理
files = os.listdir('output\\low_dim')
```

```python
out_data1 = pd.DataFrame(columns=['TPR','TNR','ReErr','RunTime','rho','sigma','n','method'])
out_data2 = pd.DataFrame(columns=['TPR','TNR','ReErr','RunTime','rho','sigma','n','method','alpha'])
```

```python
for file in files:
    data = pd.read_csv('output\\low_dim\\'+file)
    xinxi = data.iloc[0,4:]
    estimate_data = data.iloc[:,:4]
    data_mean = estimate_data.mean(axis=0)
    ## 把data_mean和xinxi合并
    data_mean = pd.DataFrame(pd.concat([data_mean,xinxi],axis=0)).T
    if 'Chen_Li' in file:
        data_mean.columns = ['TPR','TNR','ReErr','RunTime','rho','sigma','n','method','alpha']
        out_data2 = out_data2.append(data_mean,ignore_index=True)
    else:
        data_mean.columns = ['TPR','TNR','ReErr','RunTime','rho','sigma','n','method']
        out_data1 = out_data1.append(data_mean,ignore_index=True)
```

```python
out_data1.to_excel('output\\low_dim_out_data1.xlsx')
out_data2.to_excel('output\\low_dim_out_data2.xlsx')
```
